```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
```

```python
text = '''
Access to information: You can search for almost any topic and find articles, videos, and other resources online. 
Communication tools: Email, messaging apps, and video calls let you connect with people anywhere in the world. 
Social networking: Platforms like Facebook, Twitter, and Instagram allow you to connect with friends and family, share updates, and follow others. 
Online shopping: Purchase goods from stores worldwide directly through the internet. 
Digital entertainment: Watch movies, listen to music, and play games online. 
Learning and education: Access online courses, tutorials, and educational materials. 
Web browsing: Navigate different websites using a web browser to find information. 
File sharing: Send and receive files with others. 
Need for internet connection: To access the internet, you need a device like a computer, phone, or tablet and an internet service provider. 
'''
```

```python
plt.figure(figsize=(10,10))
wordcloud = WordCloud(background_color='lightblue').generate(text)
plt.imshow(wordcloud) 
plt.axis('off')
plt.show()
```
